```python
import pandas as pd

# הגדרות כלליות
file_name = 'data.xlsx'
sheets = ['בשרוני', 'ויה איטלי', 'הירוקה']

# פונקציה לעיבוד נתוני התפריט (חלק עליון של האקסל)
def process_menu(df):
    # חילוץ המנות (שורות 1-6, עמודות A-D)
    menu = df.iloc[0:5, 0:4].copy()
    menu.columns = ['dish', 'cost', 'price', 'prep_time']
    
    # המרה למספרים וחישוב רווח
    menu[['cost', 'price', 'prep_time']] = menu[['cost', 'price', 'prep_time']].apply(pd.to_numeric, errors='coerce')
    menu['profit'] = menu['price'] - menu['cost']
    return menu
#---------------------------------------------------------------------------------------------------------------
# פונקציה לעיבוד הביקושים (חלק תחתי של האקסל)
def process_demands(df, menu_items):
    # השעות נמצאות בשורה 9, הנתונים משורה 10 ומטה
    # אנחנו מחלצים את כל הנתונים החל מהעמודה השנייה (B)
    demand_raw = df.iloc[9:, 1:].copy()
    
    # חישוב ממוצע לכל עמודה
    avg_series = demand_raw.mean()
    
    # ארגון הנתונים במבנה של: מנה | שעה | ביקוש ממוצע
    demands_list = []
    hours = ['10:00', '11:00', '12:00', '13:00', '14:00', '15:00']
    
    col_idx = 0
    for dish in menu_items:
        for hour in hours:
            if col_idx < len(avg_series):
                demands_list.append({
                    'dish': dish,
                    'hour': hour,
                    'avg_demand': avg_series.iloc[col_idx]
                })
                col_idx += 1
                
    return pd.DataFrame(demands_list)
#-----------------------------------------------------------------------------------------------------------------
# הרצה על כל המסעדות ושמירת התוצאות
all_menus = {}
all_demands = {}

for sheet in sheets:
    # קריאת הלשונית
    df_temp = pd.read_excel(file_name, sheet_name=sheet, header=None)
    
    # עיבוד תפריט
    menu_res = process_menu(df_temp)
    all_menus[sheet] = menu_res
    
    # עיבוד ביקוש
    demand_res = process_demands(df_temp, menu_res['dish'].tolist())
    all_demands[sheet] = demand_res
    
    # הדפסת תוצאות לבדיקה
    print(f"\n=== ניתוח מסעדת: {sheet} ===")
    print("תפריט ורווחיות:")
    print(menu_res[['dish', 'profit', 'prep_time']])
    print("\nדוגמת ביקושים ממוצעים (לשעות הראשונות):")
    print(demand_res.head(6))
```

```python
import pulp
import pandas as pd
from IPython.display import display, HTML
import warnings

# ביטול הצגת אזהרות מיותרות של המערכת
warnings.filterwarnings('ignore', category=UserWarning)


#----------------הפונקציה מריצה מודל אופטימיזציה למסעדה אחת-----------------------

#הפונקציה מקבלת:שם מסעדה,זמני הכנה לכל מנה,רווח לכל מנה,ביקוש יומי לכל מנה  
#ומחזירה:לוח עבודה אופטימלי לפי שעות ורווח יומי אופטימלי

def solve_restaurant_optimization(restaurant_name, prep_times, profits, daily_demands):
    
    clean_name = restaurant_name.replace(" ", "_")
    prob = pulp.LpProblem(f"Optimization_{clean_name}", pulp.LpMaximize)
   
    hours = [10, 11, 12, 13, 14, 15] 
    items = list(profits.keys())

    
    # הגדרת משתני החלטה: כמות מנות מכל סוג בכל שעה (מספר שלם וחיובי) 
    x = pulp.LpVariable.dicts("Prod", (items, hours), lowBound=0, cat='Integer')
    
    # משתנה החלטה בינארי לעובד נוסף בכל שעה (1 - שוכרים, 0 - לא שוכרים)
    y = pulp.LpVariable.dicts("Extra_Worker", hours, cat='Binary')
    
    # פונקציית המטרה: מקסום סך הרווח מהמנות בניכוי עלות עובד נוסף (60 ש"ח לשעה)
    prob += pulp.lpSum([profits[i] * x[i][t] for i in items for t in hours]) - pulp.lpSum([60 * y[t] for t in hours])
    
    #:אילוצים
    
    # 1. אילוץ מגבלת עמדות מטבח בכל שעה - לכל היותר 2 עמדות (עד 120 דקות) 
    for t in hours:
        prob += pulp.lpSum([prep_times[i] * x[i][t] for i in items]) <= 120

    # 2. אילוץ מגבלת קיבולת יומית כוללת - 4 שעות בסיס (480 דקות) + 120 דקות על כל שעת עובד נוסף
    prob += pulp.lpSum([prep_times[i] * x[i][t] for i in items for t in hours]) <= 480 + 120 * pulp.lpSum([y[t] for t in hours])

    # 3. אילוץ ביקוש יומי ממוצע לכל מנה (חסם עליון) 
    for i in items:
        prob += pulp.lpSum([x[i][t] for t in hours]) <= daily_demands[i]

    # 4. אילוץ סך שעות עבודה כולל לא יעלה על 6 שעות פעילות (720 דקות)
    prob += pulp.lpSum([prep_times[i] * x[i][t] for i in items for t in hours]) <= 720

        
    # פתרון המודל 
    prob.solve(pulp.PULP_CBC_CMD(msg=0))

    
 # ריכוז התוצאות לטבלה
    results = []
    for t in hours:
        # קביעת התצוגה של עובד נוסף לפי ערך המשתנה הבינארי במודל (0 או 1)
        has_extra_worker = "כן" if pulp.value(y[t]) == 1 else "לא"
        
        row = {"שעה": f"{t}:00", "עובד נוסף": has_extra_worker} 
        used_time = sum([prep_times[i] * pulp.value(x[i][t]) for i in items])
        row["דקות עבודה"] = int(used_time)
        for i in items:
            row[i] = int(pulp.value(x[i][t]))
        results.append(row)
    
    return pd.DataFrame(results), pulp.value(prob.objective)

# נתוני המסעדות כפי שמופיעים בדו"ח 
data = {
    "בשרוני": {
        "prep": {"המבורגר": 13, "סטייק": 12, "פרגית": 10, "שניצל": 10, "כבד עוף": 15},
        "profit": {"המבורגר": 50, "סטייק": 110, "פרגית": 50, "שניצל": 45, "כבד עוף": 45},
        "demand": {"המבורגר": 20, "סטייק": 10, "פרגית": 13, "שניצל": 16, "כבד עוף": 7}
    },
    "ויה איטלי": {
        "prep": {"פיצה": 14, "פסטה": 7, "רביולי": 9, "ניוקי": 9, "ריזוטו": 8},
        "profit": {"פיצה": 66, "פסטה": 40, "רביולי": 65, "ניוקי": 57, "ריזוטו": 48},
        "demand": {"פיצה": 17, "פסטה": 14, "רביולי": 11, "ניוקי": 9, "ריזוטו": 6}
    },
    "הירוקה": {
        "prep": {"סלט הבית": 5, "מג'דרה": 7, "סלט קינואה": 7, "סלט אישי": 6},
        "profit": {"סלט הבית": 66, "מג'דרה": 40, "סלט קינואה": 65, "סלט אישי": 57},
        "demand": {"סלט הבית": 17, "מג'דרה": 17, "סלט קינואה": 17, "סלט אישי": 17}
    }
}

summary_results = []

# הרצה והצגת התוצאות
for name, d in data.items():
    df, profit = solve_restaurant_optimization(name, d['prep'], d['profit'], d['demand'])
    
    # הצגת כותרת
    display(HTML(f"<h3 style='color: #2e7d32; text-align: right;'>לוח עבודה אופטימלי: {name}</h3>"))
    display(HTML(f"<p style='text-align: right;'><b>רווח יומי כולל: {profit:,.2f} ש\"ח</b></p>"))
    
    # עיצוב הטבלה (דירוג צבעים לדקות עבודה)
    styled_df = df.style.background_gradient(subset=['דקות עבודה'], cmap='YlOrRd') \
                        .set_properties(**{'text-align': 'center'}) \
                        .set_table_styles([
                            {'selector': 'th', 'props': [('background-color', '#f2f2f2'), ('font-weight', 'bold')]},
                            {'selector': 'table', 'props': [('margin-left', 'auto'), ('margin-right', '0')]}
                        ])
    
    display(styled_df)
    summary_results.append({"מסעדה": name, "רווח יומי (ש'ח)": profit})

# טבלת סיכום סופית
display(HTML("<h3 style='color: #1565c0; text-align: right;'>ריכוז רווחים סופי</h3>"))
summary_df = pd.DataFrame(summary_results)
display(summary_df.style.bar(subset=["רווח יומי (ש'ח)"], color='#5fba7d').hide_index())
```

```python
import pandas as pd
import numpy as np

# שם קובץ האקסל המרכזי שלך
file_name = 'data.xlsx'

def analyze_restaurant_variance_from_sheet(file_name, sheet_name):
    # 1. קריאת הלשונית ללא כותרות כדי להתאים למבנה הקוד שלך
    df = pd.read_excel(file_name, sheet_name=sheet_name, header=None)
    
    # 2. חילוץ טבלת הביקוש היומית:
    # הכותרות של השעות נמצאות בשורה 8 (אינדקס 8), והנתונים מתחילים משורה 9 (אינדקס 9)
    # אנחנו מדלגים על העמודה הראשונה (אינדקס 0) שהיא עמודת התאריכים
    demand_data = df.iloc[9:, 1:].copy()
    
    # המרת כל נתוני הביקוש למספרים וניקוי שורות ריקות במידה ויש
    demand_data = demand_data.apply(pd.to_numeric, errors='coerce').dropna()
    
    # 3. חישוב סך הכל הביקוש היומי של כל המנות יחד (סכימה אופקית לכל יום)
    total_daily_demand = demand_data.sum(axis=1)
    
    # 4. חישוב המדדים הסטטיסטיים של הפיזור
    mean_demand = total_daily_demand.mean()
    std_demand = total_daily_demand.std()
    variance_demand = total_daily_demand.var()
    cv = (std_demand / mean_demand) * 100
    
    # קביעת סטטוס עמידה במבחן לפי חוקי האצבע התפעוליים
    if cv <= 20:
        status = "עומד בהצטיינות (יציב מאוד)"
    elif cv <= 35:
        status = "עומד במבחן (תנודתיות מתונה)"
    else:
        status = "לא עומד במבחן (תנודתיות גבוהה)"
        
    return {
        'מסעדה': sheet_name,
        'ממוצע יומי (Mean)': round(mean_demand, 2),
        'שונות (Variance)': round(variance_demand, 2),
        'סטיית תקן (Std)': round(std_demand, 2),
        'מקדם שונות (CV)': f"{cv:.2f}%",
        'סטטוס יציבות': status
    }

# מיפוי הלשוניות בדיוק לפי השמות בקובץ שלך
sheets = ['בשרוני', 'ויה איטלי', 'הירוקה']

# הרצת הניתוח הסטטיסטי
results = []
for sheet in sheets:
    try:
        res = analyze_restaurant_variance_from_sheet(file_name, sheet)
        results.append(res)
    except Exception as e:
        print(f"❌ שגיאה בניתוח לשונית {sheet}: {e}")

# הדפסת דוח הסיכום הסטטיסטי הסופי
if results:
    df_summary = pd.DataFrame(results)
    print("\n" + "="*70)
    print("=== טבלת סיכום ומבחן שונות לכלל המסעדות במתחם המזון ===")
    print("="*70)
    print(df_summary.to_string(index=False))
    print("="*70)
```
